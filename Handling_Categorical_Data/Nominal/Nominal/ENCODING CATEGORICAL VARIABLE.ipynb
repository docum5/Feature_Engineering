```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Feature Engineering/adult.csv')
data.head()
```

# UNTUK DATA YG BINARY 


## Cara 1 

```python
data["gender"] = data["gender"].apply(lambda val: 
                                      1 if val == "Female" else 0) # Beri label 1 jika F dan 0 lainnya

data.head()
```

## Cara 2

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data["gender"] = le.fit_transform(data["gender"])
data.head()
```

# UNTUK DATA YANG LEBIH DARI 2 VARIABEL

```python
variabel_dummies_marital_status = pd.get_dummies(data['marital-status'])
print(variabel_dummies_marital_status)
```
